```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use("fivethirtyeight")
```

```python
data = sns.load_dataset("tips")
```

```python
data.head()
```

```python
sns.relplot(x= "total_bill", y = "tip", hue = "smoker", style ="sex", size = "size", data= data, sizes = (15,200) )
```

```python
sns.boxplot(data["tip"])
```

```python
# put in the x-axis a categorical variable
sns.catplot(x = "day", y = "total_bill", kind = "box", data = data)
```

```python
sns.catplot(x = "day", y = "total_bill", kind = "box", hue = "sex", data = data)
```

```python
sns.violinplot(data["total_bill"])
```

```python
sns.catplot(x = "day", y = "total_bill", kind = "violin", hue =  "sex", data=data )
```

```python
sns.catplot(x = "smoker", y= "total_bill", hue= "sex",estimator= np.var,data=data, kind = "bar")
```

```python
# histogram work on numerical data and count plot work on catoricial data. it is a special case of the bar plot
# count for the male and female who are smoking
sns.countplot(x="sex", hue= "smoker", data=data)
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
x = flights.pivot_table(index= "year", columns = "month", values= "passengers", aggfunc= "sum")
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar = False,linewidth = 0.5,annot= True, fmt= "d")
```

Clustered map

```python
#it clustered similar rows

sns.clustermap(x,annot=True, metric = "correlation")
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris, hue = "species")
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic["age"].fillna(titanic["age"].mean(), inplace= True)
```

Distplot

```python
sns.distplot(titanic["age"])
```

```python
#Histogram
sns.distplot(titanic["age"],bins = 10, kde = False)
```

```python
#KDE show probability density function
sns.distplot(titanic["age"], hist = False)
```

```python
sns.distplot(titanic[titanic["survived"]==1]["age"])
sns.distplot(titanic[titanic["survived"]==0]["age"])
```
